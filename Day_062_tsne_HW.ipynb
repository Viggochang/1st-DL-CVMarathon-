```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter
from sklearn import manifold, datasets
from time import time
```

```python
# 設定模型與繪圖參數
n_samples = 300
n_components = 2
(fig, subplots) = plt.subplots(2, 5, figsize=(15, 6))
perplexities = [4, 6, 9, 14, 21, 30, 45, 66, 100]

# 設定S曲線資料點
X, y = datasets.make_moons(n_samples, random_state=0)

# 資料原圖
ax = subplots[0][0]
ax.set_title("Original")
ax.scatter(X[:, 0], X[:, 1], c=y)
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
plt.axis('tight')

# 不同 perplexity 之下, t-SNE流形還原圖
i = 1
for perplexity in perplexities:
    tsne = manifold.TSNE(n_components = n_components, init = "random", random_state=0, perplexity=perplexity)
    t0 = time()
    X_tsne = tsne.fit_transform(X)
    t1 = time()
    print(f"perplexity={perplexity} in {t1 - t0} sec")
    
    if i <=4:
        ax = subplots[0][i]
    elif i>=5:
        ax = subplots[1][i-5]
    ax.set_title(f"Perplexity={perplexity}")    
    ax.scatter(X_tsne[:,0], X_tsne[:,1], c=y)
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    ax.axis('tight')
    i+=1
    
plt.show()    
```
